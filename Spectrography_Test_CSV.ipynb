```python
%matplotlib inline  
import csv
import matplotlib.pyplot as plt

x = []
y = []
# Import csv file
with open("Aurorite_39.CSV", "r") as file:
    data = csv.reader(file)
    for line in data:
        if line[1] == " .":
            line[1] = "0"
        x.append(line[0])
        y.append(line[1])
```

```python
plt.plot(x,y)
plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
plt.show()
```

```python
%matplotlib inline
import ipywidgets as widgets
import pandas as pd
import csv


x = []
y = []
# Import csv file
with open("Aurorite_39.CSV", "r") as file:
    data = csv.reader(file)
    for line in data:
        if line[1] == " .":
            line[1] = "0"
        x.append(line[0])
        y.append(line[1])      
```

```python
df = pd.DataFrame(list(zip(x, y)))
df[1] = df[1].astype(float)

df.plot(x=0,y=1, xlim=(1500,30), ylim=(-1.2,0), xticks=(1500,750, 0))
```
